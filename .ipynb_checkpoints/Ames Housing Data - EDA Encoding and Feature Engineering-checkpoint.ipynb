## Visualization, EDA and Feature Engineering

Let's create some charts and plots to help us understand the data as well as help us with feature engineering.
Let's start with categorical features

```python
# Custom function for plottingraph of categraphorical features
def strip_plot(data, col_list, targraphet_var):
    """plots strip plots for all the number of columns passed"""
    # calculate the number of rows in our subplot
    nrows = int(np.ceil(len(col_list)/4))
    # calculate the number of columns in our subplot
    ncols = 4
    # define our subplot
    figraph, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20, 35))
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    
    # for loop to create each of the plots
    for i, col in enumerate(col_list):
        graph = sns.stripplot(data=data, y=col, x=targraphet_var, ax=ax[i], orient='h')
        ax[i].set_title(col, fontsize=16)
        
        # since each graph will have the feature as its title, 
        # and each graph will have 'SalePrice' as the X-axis, 
        # to keep things neater, there will be no axis title.
        graph.set(xlabel=None)
        graph.set(ylabel=None)
        
        # making the x-axis neater
        graph.set(xticks=[0, 100000, 200000, 300000, 400000, 500000, 600000])
        graph.set(xticklabels=['0k','100k','200k','300k','400k','500k','600k'])
    
    # Hide unused plots
    for i in range(nrows * ncols - len(col_list)):
        ax[-i-1].set_visible(False) # Start from -1
    
    plt.suptitle('Categoric Features vs Sale Price', fontsize=30, y=1.015)
    plt.tight_layout()
```

### Manual Encoding

```python
df['Lot Shape'] = df['Lot Shape'].map({'IR3':0, 'IR2':1, 'IR1':2, 'Reg':3})
df['Land Slope'] = df['Land Slope'].map({'Gtl':0, 'Mod':1, 'Sev':2})
df['Exter Qual'] = df['Exter Qual'].map({'Po':0, 'Fa':1, 'TA':2, 'Gd':3, 'Ex':4})
df['Exter Cond'] = df['Exter Cond'].map({'Po':0, 'Fa':1, 'TA':2, 'Gd':3, 'Ex':4})
df['Bsmt Qual'] = df['Bsmt Qual'].map({'No Basement':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
df['Bsmt Cond'] = df['Bsmt Cond'].map({'No Basement':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
df['Bsmt Exposure'] = df['Bsmt Exposure'].map({'No Basement':0, 'No':1, 'Mn':2, 'Av':3, 'Gd':4})
df['BsmtFin Type 1'] = df['BsmtFin Type 1'].map({'No Basement':0, 'Unf':1, 'LwQ':2, 'Rec':3, 'BLQ':4, 'ALQ':5, 'GLQ':6})
df['BsmtFin Type 2'] = df['BsmtFin Type 2'].map({'No Basement':0, 'Unf':1, 'LwQ':2, 'Rec':3, 'BLQ':4, 'ALQ':5, 'GLQ':6})
df['Heating QC'] = df['Heating QC'].map({'Po':0, 'Fa':1, 'TA':2, 'Gd':3, 'Ex':4})
df['Electrical'] = df['Electrical'].map({'Mix':0, 'FuseP':1, 'FuseF':2, 'FuseA':3, 'SBrkr':4})
df['Kitchen Qual'] = df['Kitchen Qual'].map({'Po':0, 'Fa':1, 'TA':2, 'Gd':3, 'Ex':4})
df['Functional'] = df['Functional'].map({'Sal':0, 'Sev':1, 'Maj2':2, 'Maj1':3, 'Mod':4, 'Min2':5, 'Min1':6, 'Typ':7})
df['Fireplace Qu'] = df['Fireplace Qu'].map({'No Fireplace':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
df['Garage Finish'] = df['Garage Finish'].map({'No Garage':0, 'Unf':1, 'RFn':2, 'Fin':3})
df['Garage Qual'] = df['Garage Qual'].map({'No Garage':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
df['Garage Cond'] = df['Garage Cond'].map({'No Garage':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
df['Paved Drive'] = df['Paved Drive'].map({'N':0, 'P':1, 'Y':2})
df['Fence'] = df['Fence'].map({'No Fence':0, 'MnWw':1, 'GdWo':2, 'MnPrv':3, 'GdPrv':4})
```

### Scatterplot for numerical features

```python
# Custom function for plotting of numeric features
# This function is exactly the same as before, just for scatterplots
def create_scatterplot(data, col_list, target_var):
    """plots scatterplots for all the columns passed."""
    nrows = int(np.ceil(len(col_list)/4))
    ncols = 4
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(14, nrows*2))
    ax = ax.ravel()
    
    for i, col in enumerate(col_list):
        g = sns.regplot(data=data, x=col, y=target_var, ax=ax[i], 
                        ci=None, color='skyblue', line_kws={'color': 'blue'})
        ax[i].set_title(col, fontsize=14)
        ax[i].set_xlabel('')
        plt.tight_layout()
    
    # Hide unused plots
    for i in range(nrows * ncols - len(col_list)):
        ax[-i-1].set_visible(False) # Start from -1
    
    plt.suptitle('Scatterplot of Numeric Features vs Sale Price', fontsize=20, y=1.015)
```

## Exploring Multi Collinearity

Before we start with feature selection, we will take a look at multicollinearity. There are some features that we would already expect to be closely related, just from their names, we will take a look at them below:

1. We would expect the Square Feet of the Basement 'components' to be very closely related to the total Basement Square Feet.
1. We would expect the 1st and 2nd Flr square feet to be very closely related to the overall square feet.

```python
plt.figure(figsize=(15,8))

# plotting sum of 'bsmt' component SF vs total bsmt SF
sns.scatterplot(x=(df['BsmtFin SF 1']+df['BsmtFin SF 2']+df['Bsmt Unf SF']),y=df['Total Bsmt SF'])
plt.xlabel('Bsmt Component SF')
plt.title('Overall Bsmt SF vs Bsmt Component SF');
```

Based on the chart above, it will be unneccesary for us to use all four of these features. We will choose to keep the `Total Bsmt SF` and drop the rest.

```python
df.drop(columns=['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF'], inplace=True)
```

```python
plt.figure(figsize=(15,8))

#similarly, plotting component SF vs 'Gr Liv Area'
sns.scatterplot(x=df['1st Flr SF']+df['2nd Flr SF']+df['Low Qual Fin SF'],y=df['Gr Liv Area'])
plt.xlabel('1st and 2nd Flr SF')
plt.title('Overall SF vs 1st,2nd Flr SF');
```

Based on the chart above, it will be unneccesary for us to use all four of these features. We will choose to keep the `Gr Live Area` and drop the rest.

```python
df.drop(columns=['1st Flr SF', '2nd Flr SF', 'Low Qual Fin SF'], inplace=True)
```

```python
feat_other.append('SalePrice')
corr_matrix = df[feat_other].corr().abs()

sol = (corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1)
                  .astype(bool))
                  .stack()
                  .sort_values(ascending=False))

# Convert to dataframe and reset multi-level index
corr_data = pd.DataFrame(sol.head(20)).reset_index()

# Rename columns
corr_data.columns = 'v1', 'v2', 'pair_corr'
```

```python
def corr_target(row):
    """returns 2 new columns of correlation score between variable 1 
    and variable 2 with the target, SalePrice"""
    row['v1_y_corr'] = df.corr()['SalePrice'][row['v1']]
    row['v2_y_corr'] = df.corr()['SalePrice'][row['v2']]
    return row
```

```python
# Create df with pairwise correlation and correlation to target
corr_data = corr_data.apply(corr_target, axis=1)
```

```python
corr_data
```

As we had previously hypothesised, we are seeing pairwise correlation here between the `Exterior 1st` and `Exterior 2nd`, and `Condition 1` and `Condition 2`.
To eliminate this, we will create an interaction column for each of these two.

Addtionally, in row 3, we see that `Roof Style_Gable` and `Roof Style_Hip` have a very high pairwise correlation.
Based on each of their correlation to `SalePrice`, we will keep `Roof Style_Hip` as it has a higher absolute value of correlation with `SalePrice`.

```python
feat_ext_1st = df.filter(like='Exterior 1st').columns.tolist()
feat_cond_1 = df.filter(like='Condition 1').columns.tolist()
df.rename(columns={'Exterior 2nd_Brk Cmn': 'Exterior 2nd_BrkComm',
                  'Exterior 2nd_CmentBd': 'Exterior 2nd_CemntBd',
                  'Exterior 2nd_Wd Shng': 'Exterior 2nd_WdShing'}, inplace=True)
```

```python
for col in feat_ext_1st: 
    ext_type = col.split('_')[1]
    df[f'Exterior_{ext_type}'] = df[f'Exterior 1st_{ext_type}']*df[f'Exterior 2nd_{ext_type}']
    df.drop(columns=[f'Exterior 1st_{ext_type}',f'Exterior 2nd_{ext_type}'], inplace=True)
    
for col in feat_cond_1: 
    cond_type = col.split('_')[1]
    if cond_type == 'RRNe':
        pass
    else:
        df[f'Condition_{cond_type}'] = df[f'Condition 1_{cond_type}']*df[f'Condition 2_{cond_type}']
        df.drop(columns=[f'Condition 1_{cond_type}',f'Condition 2_{cond_type}'], inplace=True)
```

```python
df.drop(columns=['Roof Style_Gable'], inplace=True)
```

Now that we have created the necessary interaction columns, and dropped columns where we find high correlation with other features, we are ready to plug these features back into our model, and work on creating the best model to predict our `SalePrice` values in our test data set.

Before we do that, we need to create a custom function to replicate the actions we took for our training dataset, over to our test data set.

```python
def encode_dummify_drop(train_df, test_df):
    """repeating cleaning/encoding steps taken to our test data, in the 
    SAMME ORDER as taken on our train data."""
    
    test_df['Lot Shape'] =test_df['Lot Shape'].map({'IR3':0, 'IR2':1, 'IR1':2, 'Reg':3})
    test_df['Land Slope'] =test_df['Land Slope'].map({'Gtl':0, 'Mod':1, 'Sev':2})
    test_df['Exter Qual'] =test_df['Exter Qual'].map({'Po':0, 'Fa':1, 'TA':2, 'Gd':3, 'Ex':4})
    test_df['Exter Cond'] =test_df['Exter Cond'].map({'Po':0, 'Fa':1, 'TA':2, 'Gd':3, 'Ex':4})
    test_df['Bsmt Qual'] =test_df['Bsmt Qual'].map({'No Basement':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
    test_df['Bsmt Cond'] =test_df['Bsmt Cond'].map({'No Basement':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
    test_df['Bsmt Exposure'] =test_df['Bsmt Exposure'].map({'No Basement':0, 'No':1, 'Mn':2, 'Av':3, 'Gd':4})
    test_df['BsmtFin Type 1'] =test_df['BsmtFin Type 1'].map({'No Basement':0, 'Unf':1, 'LwQ':2, 'Rec':3, 'BLQ':4, 'ALQ':5, 'GLQ':6})
    test_df['BsmtFin Type 2'] =test_df['BsmtFin Type 2'].map({'No Basement':0, 'Unf':1, 'LwQ':2, 'Rec':3, 'BLQ':4, 'ALQ':5, 'GLQ':6})
    test_df['Heating QC'] =test_df['Heating QC'].map({'Po':0, 'Fa':1, 'TA':2, 'Gd':3, 'Ex':4})
    test_df['Electrical'] =test_df['Electrical'].map({'Mix':0, 'FuseP':1, 'FuseF':2, 'FuseA':3, 'SBrkr':4})
    test_df['Kitchen Qual'] =test_df['Kitchen Qual'].map({'Po':0, 'Fa':1, 'TA':2, 'Gd':3, 'Ex':4})
    test_df['Functional'] =test_df['Functional'].map({'Sal':0, 'Sev':1, 'Maj2':2, 'Maj1':3, 'Mod':4, 'Min2':5, 'Min1':6, 'Typ':7})
    test_df['Fireplace Qu'] =test_df['Fireplace Qu'].map({'No Fireplace':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
    test_df['Garage Finish'] =test_df['Garage Finish'].map({'No Garage':0, 'Unf':1, 'RFn':2, 'Fin':3})
    test_df['Garage Qual'] =test_df['Garage Qual'].map({'No Garage':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
    test_df['Garage Cond'] =test_df['Garage Cond'].map({'No Garage':0, 'Po':1, 'Fa':2, 'TA':3, 'Gd':4, 'Ex':5})
    test_df['Paved Drive'] =test_df['Paved Drive'].map({'N':0, 'P':1, 'Y':2})
    test_df['Fence'] =test_df['Fence'].map({'No Fence':0, 'MnWw':1, 'GdWo':2, 'MnPrv':3, 'GdPrv':4})
    
    test_df = pd.get_dummies(test_df, columns=['Alley']).drop('Alley_no alley access', axis=1)
    test_df = pd.get_dummies(test_df, columns=['Mas Vnr Type']).drop('Mas Vnr Type_None', axis=1)
    test_df = pd.get_dummies(test_df, columns=['Garage Type']).drop('Garage Type_No Garage', axis=1)
    test_df = pd.get_dummies(test_df, columns=['Misc Feature']).drop('Misc Feature_None', axis=1)
    test_df = pd.get_dummies(test_df, columns=['Central Air']).drop('Central Air_Y', axis=1)
    test_df = pd.get_dummies(test_df, columns=['Heating']).drop('Heating_GasA', axis=1)
    test_df = pd.get_dummies(test_df, columns=['House Style']).drop('House Style_2Story', axis=1)
    test_df = pd.get_dummies(test_df, columns=['MS Zoning']).drop('MS Zoning_RL', axis=1)
    test_df = pd.get_dummies(test_df, columns=['Exterior 1st'])
    test_df = pd.get_dummies(test_df, columns=['Exterior 2nd'])
    test_df = pd.get_dummies(test_df, columns=['Condition 1'])
    test_df = pd.get_dummies(test_df, columns=['Condition 2'])
    test_df.drop(columns=['Bldg Type'], inplace=True)
    test_df.drop(columns=['Roof Matl'], inplace=True)

    cat_feat_remaining = [col for col in test_df.columns.tolist() if test_df[col].dtype == object]
    test_df = pd.get_dummies(test_df, columns=cat_feat_remaining, drop_first=True)
    test_df['Garage Qual + Age Garage Blt'] =test_df['Garage Qual'] +test_df['Age Garage Blt']
    test_df.drop(columns=['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF'], inplace=True)
    test_df.drop(columns=['1st Flr SF', '2nd Flr SF', 'Low Qual Fin SF'], inplace=True)
    test_df.drop(columns=['Exter Qual', 'Bsmt Qual', 'Kitchen Qual'], inplace=True)
    test_df.drop(columns=['Garage Yr Blt','Garage Finish','Garage Cars','Garage Cond',
                          'Age Garage Blt'], inplace=True)

    feat_ext_1st =test_df.filter(like='Exterior 1st').columns.tolist()
    feat_cond_2 =test_df.filter(like='Condition 2').columns.tolist()
    test_df.rename(columns={'Exterior 2nd_Brk Cmn': 'Exterior 2nd_BrkComm',
                            'Exterior 2nd_CmentBd': 'Exterior 2nd_CemntBd',
                            'Exterior 2nd_Wd Shng': 'Exterior 2nd_WdShing'}, 
                   inplace=True)

    for col in feat_ext_1st: 
        ext_type = col.split('_')[1]
        test_df[f'Exterior_{ext_type}'] =test_df[f'Exterior 1st_{ext_type}']*test_df[f'Exterior 2nd_{ext_type}']
        test_df.drop(columns=[f'Exterior 1st_{ext_type}',f'Exterior 2nd_{ext_type}'], inplace=True)

    for col in feat_cond_2: 
        cond_type = col.split('_')[1]
    if cond_type == 'RRNe':
        pass
    else:
        test_df[f'Condition_{cond_type}'] =test_df[f'Condition 1_{cond_type}']*test_df[f'Condition 2_{cond_type}']
        test_df.drop(columns=[f'Condition 1_{cond_type}',f'Condition 2_{cond_type}'], inplace=True)

    test_df.drop(columns=['Roof Style_Gable'], inplace=True)
    
    # checking to see if there are additional columns in the test dataset 
    # that don't exist in our train dataset, if so, they will be removed.
    # for columns that exist only in our train dataset, we will add them
    # to the test dataset, with it's value set to zero.
    test_col = test_df.columns
    train_col = train_df.columns
    additional_cols = [col for col in test_col if col not in train_col]
    test_df.drop(columns=additional_cols, inplace=True)
    missing_cols = [col for col in train_col if col not in test_col and col != 'SalePrice']
    
    # for loop to create missing columns in test dataset
    for col in missing_cols:
        test_df[col] = 0
    
    return test_df
```
